## Data Cleaning

```python
train_data.drop(['id','CustomerId','Surname'],axis=1,inplace=True)
test_data.drop(['id','CustomerId','Surname'],axis=1,inplace=True)
```

# Using LabelEncoder

```python
le=preprocessing.LabelEncoder()

train_data.Geography=le.fit_transform(train_data.Geography)
train_data.Gender=le.fit_transform(train_data.Gender)
```

```python
le=preprocessing.LabelEncoder()

test_data.Geography=le.fit_transform(test_data.Geography)
test_data.Gender=le.fit_transform(test_data.Gender)
```

# Using Standard Scaler

```python
ss=StandardScaler()

train_data['Balance']=ss.fit_transform(train_data[['Balance']])

train_data.head()
```

```python
test_data['Balance']=ss.fit_transform(test_data[['Balance']])

test_data.head()
```

# Training and testing the model

```python
x=train_data.drop(['Exited'],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,train_data['Exited'],test_size=0.3)
```

```python
x_train.head()
```

```python
rf = RandomForestClassifier(n_estimators=100,random_state=45)
rf.fit(x_train,y_train)
```

```python
y_pred_test = rf.predict(x_test)
```

```python
accuracy_score = accuracy_score(y_test,y_pred_test)
print("accuracy score:",accuracy_score)
```
