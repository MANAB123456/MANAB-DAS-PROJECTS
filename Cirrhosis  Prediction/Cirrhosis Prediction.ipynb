```python
data_train.drop(['N_Days'],axis=1,inplace=True)
data_test.drop(['N_Days'],axis=1,inplace=True)  
```

```python
columns=['Sex','Ascites','Hepatomegaly','Spiders','Edema','Drug']
le=LabelEncoder()

data_train['Status']=le.fit_transform(data_train['Status'])

for i in columns:
    data_train[i]=le.fit_transform(data_train[i])
    data_test[i]=le.fit_transform(data_test[i])
```

```python
sc=StandardScaler()
columns_to_scale=['Age','Cholesterol','Albumin','Copper','Alk_Phos','SGOT','Tryglicerides','Platelets','Prothrombin','Stage']
data_train[columns_to_scale]=sc.fit_transform(data_train[columns_to_scale])
data_test[columns_to_scale]=sc.transform(data_test[columns_to_scale])
```

```python
from sklearn.preprocessing import RobustScaler, QuantileTransformer
import pandas as pd

# Assuming 'train' and 'test' are your dataframes
columns_to_scale = ['Age', 'Cholesterol', 'Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin', 'Stage']

# Robust Scaling
robust_scaler = RobustScaler()
data_train[columns_to_scale] = robust_scaler.fit_transform(data_train[columns_to_scale])
data_test[columns_to_scale] = robust_scaler.transform(data_test[columns_to_scale])

from sklearn.preprocessing import QuantileTransformer

quantile_transformer = QuantileTransformer(output_distribution='normal')
data_train[columns_to_scale] = quantile_transformer.fit_transform(data_train[columns_to_scale])
data_test[columns_to_scale] = quantile_transformer.transform(data_test[columns_to_scale])
```

```python
data_train.head()
```

```python
x=data_train.drop('Status',axis=1)
y=data_train['Status']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
rf_classifier = RandomForestClassifier(
    max_depth=10,
    min_samples_leaf=2,
    min_samples_split=5,
    n_estimators=200
)

rf_classifier.fit(x_train, y_train)
```

```python
y_pred = rf_classifier.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print("Accuracy:",acc)
cr=classification_report(y_test, y_pred)
print("Classification report:\n",cr)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Assuming x_train and x_test are your feature matrices and y_train, y_test are your target variables

# Define the neural network model
model = Sequential()
model.add(Dense(64, input_dim=x_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # For binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```
